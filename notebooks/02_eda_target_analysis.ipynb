```python
# Code: Target Distribution

import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
DATA_PATH = "../data/raw/application_train.csv"
df = pd.read_csv(DATA_PATH)

# Absolute counts
target_counts = df["TARGET"].value_counts().sort_index()
print("Absolute counts:", target_counts)

# Percentage Distribution
target_pct = df["TARGET"].value_counts(normalize=True).sort_index() * 100
print("Percentage Distribution:", target_pct)

```

```python
# Visualization (Simple & Clean)
plt.figure()
target_counts.plot(kind="bar")
plt.title("Target Variable Distribution")
plt.xlabel("TARGET (0 = Repaid, 1 = Default)")
plt.ylabel("Number of Applications")
plt.show()

print(f"Default Rate: {target_pct[1]:.2f}%")

```

## Target Variable Analysis

The target variable is highly imbalanced:

- ~92% of applicants repaid their loans (TARGET = 0)
- ~8% of applicants defaulted (TARGET = 1)

### Implications
- Accuracy is not a reliable evaluation metric
- ROC-AUC is more appropriate as it evaluates ranking quality
- Models must be robust to class imbalance
- False negatives (predicting non-default for a defaulter) are more costly from a business perspective

This imbalance strongly influences model selection and evaluation strategy.
