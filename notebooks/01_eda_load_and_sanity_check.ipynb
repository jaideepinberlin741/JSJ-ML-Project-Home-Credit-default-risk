```python
import pandas as pd
import numpy as np

# Display settings
pd.set_option("display.max_columns", 200)
pd.set_option("display.float_format", "{:.2f}".format)

# Load dataset
DATA_PATH = "../data/raw/application_train.csv"
df = pd.read_csv(DATA_PATH)

# Basic shape
print("Dataset shape:", df.shape)

# Preview
df.head()
```

```python
# Check 1: Target Column

assert "TARGET" in df.columns, "TARGET column is missing"
df["TARGET"].value_counts(normalize=True)
```

```python
# Check 2: Primary Key Uniqueness

df["SK_ID_CURR"].nunique() == df.shape[0]
```

```python
# Check 3: Data Types Overview

print(df.dtypes.value_counts())

print(df.select_dtypes(include=["object"]).columns[:10])
```

```python
# Check 4: Missing Values (High-Level Only)

missing_pct = df.isnull().mean() * 100
missing_pct.describe()
```

```python
#Check 5: Basic Statistical Sanity

df.describe().T.head(10)
```

- Dataset loaded successfully with ~307k records

- One row per applicant confirmed via SK_ID_CURR
